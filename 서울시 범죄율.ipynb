## ▶Part3. 결측치 이해하기_교재없음

- [정답제공] 결측치 비율 계산
- [미션3-2] 결측치 시각화 
- [미션3-3] 결측치 처리

```python
# [미션 3-1] 결측치 갯수확인해서 dfNa 변수에 할당
dfNa=df.isna().sum()
dfNa
```

```python
# [정답제공] 한개의 자료만 결측치 비율을 보고자할때는


print(dfNa.loc['살인 발생']/len(df))   # loc를 이용하여서 인덱스 이름을 직접 불러도 됨
print(dfNa[5]/len(df))  

        
```

```python
# [정답제공] Na값을 모두 확인하고자 할때는
isNaCount=df.isna().sum()
print(isNaCount)
print(isNaCount.sum())

print('--------------------------------------------------------')
if isNaCount.sum()==0:
        print('결측치 없음.')
else:
    for i,j in zip(isNaCount,isNaCount.index):
         if i>0:
            tmp=i/len(df)
            print('df자료의 NA값의 field는 %s 이며, 갯수는 %d, 전체비율은 %.2f 입니다.'%(j,i,tmp*100))
```

```python
# [정답제공] Na값 확인하는 함수로 작성
# 이러한 함수는 000.py로 저장하여서 모듈로 사용하면 편리할수 있음.
# https://wikidocs.net/29

def nacheck(df):
    isNaCount=df.isna().sum()
    if isNaCount.sum()==0:
            print('결측치 없음.')
    else:
        for i,j in zip(isNaCount,isNaCount.index):
             if i>0:
                tmp=i/len(df)
                print('df자료의 NA값의 field는 %s 이며, 갯수는 %d, 전체비율은 %.2f 입니다.'%(j,i,tmp*100))
nacheck(df)
```

```python
# [미션3-2] 히트맵으로 결측치 시각화
import seaborn as sns
sns.heatmap(df.isnull(),cbar=False)
```

```python
# [미션3-3] 결측치가 있는 행 0으로 변환
# 주의: df=결측치 0으로 변환   해야함.   그냥 결측치0으로 변환만하면 화면에만 보이는것이고 실제 자료는 변동사항 없음.
df=df.fillna(0)
```

```python
# [미션3-3] 히트맵으로 결측치 시각화함
nacheck(df) 
sns.heatmap(df.isnull(),cbar=False)
```
